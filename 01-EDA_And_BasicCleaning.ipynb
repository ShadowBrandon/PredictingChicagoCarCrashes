```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime 
%matplotlib inline

data = pd.read_csv('Original_Traffic_Crashes_-_Crashes.csv')
```

# Removing Unncessary Variables:

```python
data = data.drop(['MOST_SEVERE_INJURY','INJURIES_NON_INCAPACITATING','INJURIES_INCAPACITATING','RD_NO','CRASH_DATE_EST_I','TRAFFIC_CONTROL_DEVICE','DEVICE_CONDITION','FIRST_CRASH_TYPE','TRAFFICWAY_TYPE','LANE_CNT','NUM_UNITS','CRASH_HOUR','CRASH_DAY_OF_WEEK','CRASH_MONTH'], axis =1)
```

```python
data = data.drop(['ALIGNMENT','REPORT_TYPE','CRASH_TYPE','INTERSECTION_RELATED_I','PHOTOS_TAKEN_I','STATEMENTS_TAKEN_I','DOORING_I','WORK_ZONE_I','WORK_ZONE_TYPE','WORKERS_PRESENT_I'], axis =1)
```

```python
data = data.drop(['ROAD_DEFECT','NOT_RIGHT_OF_WAY_I','HIT_AND_RUN_I','DAMAGE','DATE_POLICE_NOTIFIED','PRIM_CONTRIBUTORY_CAUSE','SEC_CONTRIBUTORY_CAUSE'], axis =1)
```

```python
data = data.drop(['ROADWAY_SURFACE_COND','STREET_NO','STREET_DIRECTION','STREET_NAME'], axis =1)
```

```python
data = data.drop(['INJURIES_REPORTED_NOT_EVIDENT','INJURIES_NO_INDICATION','INJURIES_UNKNOWN'], axis =1)
```

```python
data = data.drop(['LONGITUDE','LATITUDE'], axis =1)
data = data.drop(['LOCATION'], axis =1)
```

```python
data['CRASH_DATE'] = pd.to_datetime(data['CRASH_DATE'])
data['COUNTER'] = (range(len(data)))
```

```python
data = data[data.notnull()]
```

### Visualizations Before Cleaning

```python
data_EDA=data
data_EDA.groupby(['YEAR','CRASH_MONTH']).count()['COUNTER'][:-1].plot(figsize = (15,6), title = 'Crash', color='#6699cc')
plt.ylabel('Number of Requests');
```

```python
# create aggregations
day_of_week_count = data[['CRASH_DAY_OF_WEEK', 'COUNTER']].groupby('CRASH_DAY_OF_WEEK').count()['COUNTER']
year_count = data[(data['YEAR']<=2019)][['YEAR', 'COUNTER']].groupby('YEAR').count()['COUNTER']
month_count = data[['CRASH_MONTH', 'COUNTER']].groupby('CRASH_MONTH').count()['COUNTER']
hour_count = data[['CRASH_HOUR', 'COUNTER']].groupby('CRASH_HOUR').count()['COUNTER']

# create figure
plt.figure(figsize=(15,9)).tight_layout()

# day of week plot
plt.subplot(2,2,1)
plt.bar(day_of_week_count.index, day_of_week_count,color='#6699cc')
plt.ylabel('Count of Crash')
plt.xlabel('Day of Week')
plt.title('Number of Crash by Day of Week')

# year plot
plt.subplot(2,2,2)
plt.bar(year_count.index, year_count,color='#6699cc')
plt.ylabel('Count of Crash')
plt.xlabel('Year')
plt.title('Number of Crash by Year')

# month plot
plt.subplot(2,2,3)
plt.bar(month_count.index, month_count,color='#6699cc')
plt.ylabel('Count of Crash')
plt.xlabel('Month')
plt.title('Number of Crash by Month')

# hour plot
plt.subplot(2,2,4)
plt.bar(hour_count.index, hour_count,color='#6699cc')
plt.ylabel('Count of Crash')
plt.xlabel('Hour')
plt.title('Number of Crash by Hour')
plt.tight_layout();

```

```python
#LOOKING AT ONLY 2017 DATA

# create aggregations
data_2017= data[(data['YEAR']==2017)]
day_of_week_count_2017 = data_2017[['CRASH_DAY_OF_WEEK', 'COUNTER']].groupby('CRASH_DAY_OF_WEEK').count()['COUNTER']
year_count_2017 = data_2017[(data_2017['YEAR']==2017)][['YEAR', 'COUNTER']].groupby('YEAR').count()['COUNTER']
month_count_2017 = data_2017[['CRASH_MONTH', 'COUNTER']].groupby('CRASH_MONTH').count()['COUNTER']
hour_count_2017 = data_2017[['CRASH_HOUR', 'COUNTER']].groupby('CRASH_HOUR').count()['COUNTER']

# create figure
plt.figure(figsize=(15,9)).tight_layout()

# day of week plot
plt.subplot(2,2,1)
plt.bar(day_of_week_count_2017.index, day_of_week_count_2017,color='#6699cc')
plt.ylabel('Count of Crash')
plt.xlabel('Day of Week')
plt.title('Number of Crash by Day of Week')

# year plot
plt.subplot(2,2,2)
plt.bar(year_count_2017.index, year_count_2017,color='#6699cc')
plt.ylabel('Count of Crash')
plt.xlabel('Year')
plt.title('Number of Crash by Year')

# month plot
plt.subplot(2,2,3)
plt.bar(month_count_2017.index, month_count_2017,color='#6699cc')
plt.ylabel('Count of Crash')
plt.xlabel('Month')
plt.title('Number of Crash by Month')

# hour plot
plt.subplot(2,2,4)
plt.bar(hour_count_2017.index, hour_count_2017,color='#6699cc')
plt.ylabel('Count of Crash')
plt.xlabel('Hour')
plt.title('Number of Crash by Hour')
plt.tight_layout();

```
